```python
import ast
import pandas as pd
import pickle
import numpy as np

def parse_list(x):
    try:
        return ast.literal_eval(x)
    except (SyntaxError, ValueError):
        return x
```

# netflix_all

```python
df = pd.read_csv('netflix_all.csv')
    
for column in ['imdb_genres', 'tmdb_genres', 'imdb_cast', 'tmdb_cast']:
    df[column] = df[column].apply(parse_list)

df
```

```python
# to keep the 7-digit format, imdb_cast is saved as string

df['imdb_cast'][0]
```

# netflix_imdb

```python
df = pd.read_csv('netflix_imdb.csv')

for column in ['imdb_ids', 'imdb_titles', 'imdb_years', 'imdb_scores', 'tmdb_ids']:
    df[column] = df[column].apply(parse_list)
    
df
```

# netflix_tmdb

```python
df = pd.read_csv('netflix_tmdb.csv')

for column in ['tmdb_ids', 'tmdb_titles', 'tmdb_years', 'tmdb_scores', 'imdb_ids']:
    df[column] = df[column].apply(parse_list)
    
df
```

# netflix_imdb_tmdb

```python
df = pd.read_csv('netflix_imdb_tmdb.csv')

for column in ['imdb_year', 'tmdb_year']:
    for i in range(len(df)):
        if not np.isnan(df[column][i]):
            df.at[i, column] = int(df[column][i])

df
```

```python
df_2 = df.loc[~np.isnan(df[column]), column].astype(int)
df_2 = pd.DataFrame(df_2)
```
